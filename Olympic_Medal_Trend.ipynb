```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

%matplotlib notebook

year_1980 = pd.read_html("https://en.wikipedia.org/wiki/1980_Summer_Olympics_medal_table")[2]
year_1984 =pd.read_html("https://en.wikipedia.org/wiki/1984_Summer_Olympics_medal_table")[2]
year_1988=pd.read_html("https://en.wikipedia.org/wiki/1988_Summer_Olympics_medal_table")[1]
year_1992 = pd.read_html("https://en.wikipedia.org/wiki/1992_Summer_Olympics_medal_table")[2]
year_1996=pd.read_html("https://en.wikipedia.org/wiki/1996_Summer_Olympics_medal_table")[2]
year_2000=pd.read_html("https://en.wikipedia.org/wiki/2000_Summer_Olympics_medal_table")[2]
year_2004 = pd.read_html("https://en.wikipedia.org/wiki/2004_Summer_Olympics_medal_table")[2]
year_2008=pd.read_html("https://en.wikipedia.org/wiki/2008_Summer_Olympics_medal_table")[2]
year_2012=pd.read_html("https://en.wikipedia.org/wiki/2012_Summer_Olympics_medal_table")[2]
year_2016=pd.read_html("https://en.wikipedia.org/wiki/2016_Summer_Olympics_medal_table")[2]
year_2020=pd.read_html("https://en.wikipedia.org/wiki/2020_Summer_Olympics_medal_table")[2]

##data cleaning

year_1980['Nation'] = year_1980["Nation"].str.replace(r"\s*\(\w*\)","")
year_1984['Nation'] = year_1984["Nation"].str.replace(r"\s*\(\w*\)","")
year_1988['Nation'] = year_1988["Nation"].str.replace(r"\s*\(\w*\)","")
year_1992['Nation'] = year_1992["Nation"].str.replace(r"\s*\(\w*\)","")
year_1996['Nation'] = year_1996["Nation"].str.replace(r"\s*\(\w*\)","")
year_2000['Nation'] = year_2000["Nation"].str.replace(r"\s*\(\w*\)","")
year_2004['Nation'] = year_2004["Nation"].str.replace(r"\s*\(\w*\)","")
year_2008['NOC'] = year_2008["NOC"].str.replace(r"\s*\(\w*\)","")
year_2012['NOC'] = year_2012["NOC"].str.replace(r"\s*\(\w*\)","")
year_2016['NOC'] = year_2016["NOC"].str.replace(r"\s*\(\w*\)","")
year_2020['Team'] = year_2020["Team"].str.replace(r"\s*\(\w*\)","")

year_1980['Nation'] = year_1980["Nation"].str.replace(r"\*","")
year_1984['Nation'] = year_1984["Nation"].str.replace(r"\*","")
year_1988['Nation'] = year_1988["Nation"].str.replace(r"\*","")
year_1992['Nation'] = year_1992["Nation"].str.replace(r"\*","")
year_1996['Nation'] = year_1996["Nation"].str.replace(r"\*","")
year_2000['Nation'] = year_2000["Nation"].str.replace(r"\*","")
year_2004['Nation'] = year_2004["Nation"].str.replace(r"\*","")
year_2008['NOC'] = year_2008["NOC"].str.replace(r"\*","")
year_2012['NOC'] = year_2012["NOC"].str.replace(r"\*","")
year_2016['NOC'] = year_2016["NOC"].str.replace(r"\*","")
year_2020['Team'] = year_2020["Team"].str.replace(r"\*","")

## Selecting required data for study

nations = ['India', "United States" ]
year_1980 = year_1980[year_1980['Nation'].isin(nations)]

year_1984 = year_1984[year_1984['Nation'].isin(nations)]
year_1988 = year_1988[year_1988['Nation'].isin(nations)]
year_1992 = year_1992[year_1992['Nation'].isin(nations)]
year_1996 = year_1996[year_1996['Nation'].isin(nations)]
year_2000 = year_2000[year_2000['Nation'].isin(nations)]
year_2004 = year_2004[year_2004['Nation'].isin(nations)]
year_2008 = year_2008[year_2008['NOC'].isin(nations)]
year_2012 = year_2012[year_2012['NOC'].isin(nations)]
year_2016 = year_2016[year_2016['NOC'].isin(nations)]
year_2020 = year_2020[year_2020['Team'].isin(nations)]

##manipulating dataset

year_1980.insert(1,"Year" , '1980')
year_1984.insert(1,"Year" , '1984')
year_1988.insert(1,"Year" , '1988')
year_1992.insert(1,"Year" , '1992')
year_1996.insert(1,"Year" , '1996')
year_2000.insert(1,"Year" , '2000')
year_2004.insert(1,"Year" , '2004')
year_2008.insert(1,"Year" , '2008')
year_2012.insert(1,"Year" , '2012')
year_2016.insert(1,"Year" , '2016')
year_2020.insert(1,"Year" , '2020')

year_2008 = year_2008.rename(columns = {'NOC' : 'Nation'})
year_2012 = year_2012.rename(columns = {'NOC' : 'Nation'})
year_2016 = year_2016.rename(columns = {'NOC' : 'Nation'})
year_2020 = year_2020.rename(columns = {'Team' : 'Nation'})

year_1980 = year_1980.drop(columns={'Rank' , 'Total'})
year_1980 = year_1980.set_index('Year')
year_1984 = year_1984.drop(columns={'Rank' , 'Total'}).set_index('Year')
year_1988 = year_1988.drop(columns={'Rank','Total' }).set_index('Year')
year_1992 = year_1992.drop(columns={'Rank','Total'}).set_index('Year')
year_1996 = year_1996.drop(columns={'Rank','Total'}).set_index('Year')
year_2000 = year_2000.drop(columns={'Rank','Total'}).set_index('Year')
year_2004 = year_2004.drop(columns={'Rank' ,'Total'}).set_index('Year')
year_2008 = year_2008.drop(columns={'Rank','Total' }).set_index('Year')
year_2012 = year_2012.drop(columns={'Rank','Total' }).set_index('Year')
year_2016 = year_2016.drop(columns={'Rank','Total'}).set_index('Year')
year_2020 = year_2020.drop(columns={'Rank','Total'}).set_index('Year')

main1 = pd.concat([year_1980 , year_1984])
frames = [main1 , year_1988 , year_1992 , year_1996 , year_2000 , year_2004 , year_2008 , year_2012 , year_2016 , year_2020]
main = pd.concat(frames)

India = main[main['Nation'] == 'India']
United_States = main[main['Nation'] == 'United States']

India = India.reset_index()
rows = {'Year' : ['1984' , '1988' , '1992'],
 'Nation' : ['India' , 'India' , 'India'],
 'Gold' : ['0' , '0','0'],
 'Silver' :['0' ,'0','0'],
 'Bronze' :['0','0','0'] }

                    
India_df = pd.DataFrame(rows)

#India = India.append(row1 , row2, row3 , ignore_index=True)
India = pd.concat([India , India_df])
India['Year'] = India["Year"].astype(str)
India = India.sort_values(by = ['Year'])
India['Year'] = India['Year'].astype(int)
India['Gold'] = India['Gold'].astype(int)
India['Silver'] = India['Silver'].astype(int)
India['Bronze'] = India['Bronze'].astype(int)
India = India.set_index("Year")
India = India.drop(columns = {'Nation' })

United_States = United_States.drop(columns = {'Nation'})
United_States = United_States.reset_index()
df2 = {'Year': 1980, 'Gold': 0, 'Silver': 0 , 'Bronze' : 0}
United_States = United_States.append(df2, ignore_index = True)
United_States['Year'] = United_States['Year'].astype(str)
United_States = United_States.sort_values(by = 'Year')
United_States = United_States.set_index('Year')
```

```python
India.plot(kind='bar' , stacked = 'true' , figsize = (8,5))
plt.xticks(rotation = 0)
plt.legend(['Gold Medal','Silver Medal','Bronze Medal'] , title = "Medal Category" )
plt.title("Olympic Medal Trend - India (1980 - 2020)")
plt.xlabel("Summer Olympics held(years)")
plt.ylabel("Number of Medals won")

totals = India.sum(axis=1)

for x, y in enumerate(totals):
    #if y != 0:
    plt.text(x, y, y, ha = 'center', va='bottom')

plt.gca().yaxis.grid(which = 'major', color = '#e1e1e1', linestyle = '--', linewidth = 0.5)       
```
